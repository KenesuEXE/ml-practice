## Data Visualization using Seaborn



### Line Charts

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

spotify_df = pd.read_csv('csv_files/spotify.csv', index_col="Date", parse_dates=True)
spotify_df.head()
```

```python
plt.figure(figsize=(14,6))
plt.title("Daily Global Streams of Popular Songs in 2017-2018")
sns.lineplot(data=spotify_df)
```

### Bar Charts

```python
flight_delays_df = pd.read_csv('csv_files/flight_delays.csv', index_col="Month")
flight_delays_df.head()
```

```python
plt.figure(figsize=(10,6))
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")
plt.ylabel("Arrival delay (min)")
sns.barplot(x=flight_delays_df.index, y=flight_delays_df['NK'])
```

```python
ign_scores_df = pd.read_csv('csv_files/ign_scores.csv', index_col='Platform')
ign_scores_df.head()
```

```python
plt.figure(figsize=(16,9))
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.title('Average Scores of Racing Games per Platform')
sns.barplot(x=ign_scores_df.index, y=ign_scores_df['Racing'])
```

### Heat Maps

```python
plt.figure(figsize=(16,9))
plt.title("Average Arrival Delay for Each Airline")
sns.heatmap(data=flight_delays_df, annot=True)
```

```python
plt.figure(figsize=(14,7))
plt.title("Average Scores per Game Genre and Platform")
sns.heatmap(data=ign_scores_df, annot=True)
```

### Scatter Plots

```python
insurance_df = pd.read_csv('csv_files/insurance.csv')
insurance_df.head()
```

```python
sns.scatterplot(x=insurance_df['bmi'], y=insurance_df['charges'])
```

```python
sns.regplot(x=insurance_df['bmi'], y=insurance_df['charges'])
```

```python
sns.scatterplot(x=insurance_df['bmi'], y=insurance_df['charges'], hue=insurance_df['smoker'])
# Use hue if third variable is limited to 1 or 2 unique values (high or low, true or false)
```

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_df)
```

```python
sns.swarmplot(x=insurance_df['smoker'],
              y=insurance_df['charges'])
```

```python
candy_df = pd.read_csv('csv_files/candy.csv', index_col='id')
candy_df.head()
```

```python
# Popularity vs Candy content: no correl
sns.scatterplot(x=candy_df['sugarpercent'], y=candy_df['winpercent'])
```

```python
sns.regplot(x=candy_df['sugarpercent'], y=candy_df['winpercent'])
```

```python
# Popularity vs Price vs Chocolate content
sns.scatterplot(x=candy_df['pricepercent'], y=candy_df['winpercent'], hue=candy_df['chocolate'])
```

```python
# Chocolate: expensive = popular, Non-Chocolate cheaper = popular
sns.lmplot(x='pricepercent', y='winpercent', hue='chocolate', data=candy_df)
```

```python
# Popularity vs Chocolate: chocolate = more popular
sns.swarmplot(x=candy_df.chocolate, y=candy_df.winpercent)
```

### Histograms

```python
iris_df = pd.read_csv('csv_files/iris.csv', index_col="Id")
iris_df.head()
```

```python
sns.histplot(iris_df['Petal Length (cm)'])
```

```python
sns.histplot(data=iris_df, x='Petal Length (cm)', hue='Species')
plt.title("Histogram of Petal Lengths, by Species")
```

```python
cancer_df = pd.read_csv('csv_files/cancer.csv')
cancer_df.head()
```

```python
sns.histplot(data=cancer_df, x='Area (mean)', hue='Diagnosis')
```

### Density Plots

```python
# kernel density estimate (KDE)
sns.kdeplot(data=iris_df['Petal Length (cm)'], fill=True)
```

```python
# 2D KDE plot
sns.jointplot(x=iris_df['Petal Length (cm)'], y=iris_df['Sepal Width (cm)'], kind="kde")
```

```python
sns.kdeplot(data=iris_df, x='Petal Length (cm)', hue='Species', fill=True)
plt.title("Distribution of Petal Lengths, by Species")
```

```python
sns.kdeplot(data=cancer_df, x='Radius (worst)', hue='Diagnosis')
```
